## Problem 3: Is the average of polls better than just one poll?

#### Problem 3(a)

Most undecided voters vote for one of the two candidates at the election. Therefore, the reported percentages underestimate the final value of both candidates. However, if we assume the undecided will split evenly, then the observed difference should be an unbiased estimate of the final difference. 

Add a new column to the `election` DataFrame containg the difference between Obama and Romeny called `Diff`. 

```python
#your code here
url = 'http://elections.huffingtonpost.com/pollster/2012-general-election-romney-vs-obama.csv'
s = requests.get(url).content
election = pd.read_csv(StringIO(s.decode('utf-8')))
#election.head()
election['Diff'] = election['Obama']/100 - election['Romney']/100
election.head()
```

#### Problem 3(b)

Make a plot of the differences for the week before the election (e.g. 5 days) where the days are on the x-axis and the differences are on the y-axis.  Add a horizontal line showing 3.9%: the difference between Obama and Romney on election day.

```python
#your code here
election3b = election[(election['End Date'] >= '2012-10-29') & (election['End Date'] <= '2012-11-02')]


#Plot
plt.scatter(pd.to_datetime(election3b['End Date']).values, election3b['Diff'].values)
plt.axhline(y=0.039)
plt.xticks(rotation = 90)
plt.show()
```

#### Problem 3(c) 

Make a plot showing the differences by pollster where the pollsters are on the x-axis and the differences on the y-axis. 

```python
#your code here
#your code here
pollsters = list(set(election3b['Pollster'].values))
pollsters3c = {}
for i in range(len(pollsters)):
    pollsters3c[pollsters[i]] = i
print(pollsters3c)
```

```python
pollValues = [pollsters3c[pollsters] for pollsters in election3b['Pollster'].values]
plt.scatter(pollValues, election3b['Diff'].values)
plt.xticks(rotation = 90)
plt.xticks(pollValues, election3b['Pollster'].values)
plt.axhline(y=0)
#plt.show()

plt.title('Obama vs. Romney')
plt.ylabel('The difference in percentage')
plt.show()
```

#### Problem 3(d)

Take the average for each pollster and then compute the average of that. Given this difference how confident would you have been of an Obama victory?

**Hint**: Compute an estimate of the SE of this average based exclusively on the observed data. 

```python
#your code here
election3d = election.loc[:,['Pollster','Diff']].copy()
election3d.head()
```

```python
election3dAvg = election3d.groupby(election3d.Pollster).mean()
election3dAvg.head()
```

```python
print('The average of the pollsters: ')
print(np.mean(election3dAvg['Diff'].values))
print('The standard deviation: ')
print(np.std(election3dAvg['Diff'].values))
```

Answer: 

The election could of gone either way when compared to the percentage in a 2.5% in Obama's favor.
